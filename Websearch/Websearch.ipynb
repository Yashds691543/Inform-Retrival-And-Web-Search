```python
import requests
from bs4 import BeautifulSoup
import re

def get_unique_words_and_frequency(words):
    temp = {}
    for word in sorted(set(words)):
        temp[word] = words.count(word)
    return temp

def get_all_words(url, sample_size=40):
    html_code_1 = requests.get(url, verify=False).text
    data = BeautifulSoup(html_code_1, 'html.parser')
    
    links = []
    for link in data.find_all("a"):
        href = link.get("href")
        if href is not None:
            if href.startswith("http"):
                links.append(href)
            else:
                full_url = requests.compat.urljoin(url, href)
                links.append(full_url)
    
    for link in links:
        try:
            html_code = requests.get(link, verify=False).text
            code_to_text = BeautifulSoup(html_code, 'html.parser').get_text()
            match_pattern = re.compile(r"\b[a-zA-Z]+\b")
            wordsList = list(map(str.lower, match_pattern.findall(code_to_text)))
            vocab_with_freq = get_unique_words_and_frequency(wordsList)
            
            print(f"Link address: {link}")
            print("Total number of clean words =", len(wordsList))
            print("=================================")
            print(" Word count ")
            print("=================================")
            for word, freq in vocab_with_freq.items():
                print(f"{word}: {freq}")
            print("\n\n")
        except Exception as e:
            print(f"Failed to process {link}: {e}")
            continue

if __name__ == '__main__':
    course_website = "https://www.cs.memphis.edu/~vrus/teaching/ir-websearch/"
    get_all_words(course_website)
```
